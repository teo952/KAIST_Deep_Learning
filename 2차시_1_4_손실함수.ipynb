```python
import torch
import torch.nn as nn

# 예제: 3개의 클래스 중 하나를 분류하는 문제
# 모델 출력 (각 클래스의 예측 확률, 배치 크기 1)
outputs = torch.tensor([[0.2, 0.5, 0.3]]) # 모델이 예측한 확률 분포
labels = torch.tensor([1]) # 실제 정답 (클래스 1)

# 교차 엔트로피 손실 함수 정의
criterion = nn.CrossEntropyLoss()

# 손실 값 계산
loss = criterion(outputs, labels)

print(f"손실 값: {loss.item():.4f}")

```
