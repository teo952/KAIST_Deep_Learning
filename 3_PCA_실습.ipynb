```python
import numpy as np
import matplotlib.pyplot as plt

# 키(Height)와 몸무게(Weight) 데이터 생성
data = np.array([
    [160, 50],
    [165, 55],
    [170, 60],
    [175, 65],
    [180, 70]
])

# 데이터 시각화
plt.scatter(data[:, 0], data[:, 1], color='blue', label='Original Data')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Original Data')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

# PCA 모델 생성 (1차원으로 축소)
pca = PCA(n_components=1)
data_pca = pca.fit_transform(data)

# 변환된 데이터 출력
print("Original Data:")
print(data)
print("\nPCA Reduced Data:")
print(data_pca)
```

```python
# PCA 변환된 데이터 복원
data_reconstructed = pca.inverse_transform(data_pca)

# 데이터 시각화
plt.scatter(data[:, 0], data[:, 1], color='blue', label='Original Data')
plt.scatter(data_reconstructed[:, 0], data_reconstructed[:, 1],
            color='red', label='PCA Reduced Data')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('PCA Transformation')
plt.legend()
plt.show()
```
