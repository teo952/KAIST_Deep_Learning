```python
import numpy as np
import matplotlib.pyplot as plt

# 3   ( 50) 
np.random.seed(42)
group1 = np.random.normal(loc=[5, 5, 5], scale=1, size=(50, 3))
group2 = np.random.normal(loc=[0, 0, 0], scale=1, size=(50, 3))
group3 = np.random.normal(loc=[-5, -5, -5], scale=1, size=(50, 3))

#   
data = np.vstack([group1, group2, group3])

#   (3D)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2],
           c=['r'] * 50 + ['g'] * 50 + ['b'] * 50)
ax.set_title('3D Data Visualization')
plt.show()
```

```python
from sklearn.manifold import TSNE

# t-SNE 
tsne = TSNE(n_components=2, random_state=42)
data_2d = tsne.fit_transform(data)

#   
print("Original Data Shape:", data.shape)
print("Reduced Data Shape:", data_2d.shape)
```

```python
# t-SNE  
plt.figure(figsize=(8, 6))
plt.scatter(data_2d[:50, 0], data_2d[:50, 1], color='red', label='Group 1')
plt.scatter(data_2d[50:100, 0], data_2d[50:100, 1], color='green', label='Group 2')
plt.scatter(data_2d[100:, 0], data_2d[100:, 1], color='blue', label='Group 3')
plt.title("t-SNE Visualization (2D)")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.legend()
plt.grid()
plt.show()
```
